```python
!pip install opendatasets --upgrade --quiet
import opendatasets as od
dataset_url = 'https://www.kaggle.com/datasets/analystanand/speaker-recognition/data'
od.download(dataset_url)
```

```python
import os
import pandas as pd
data_dir = './rossmann-store-sales'
os.listdir(data_dir)
train_csv = data_dir + '/train.csv'
ross_df = pd.read_csv(train_csv, low_memory=False)

store_csv = data_dir + '/store.csv'
store_df = pd.read_csv(store_csv)

test_csv = data_dir + '/test.csv'
test_df = pd.read_csv(test_csv)
```

```python
mege_df= ross_df.merge(store_df, how='left', on='Store')
mege_test_df= test_df.merge(store_df, how='left', on='Store')
```

```python
mege_df['Date']=pd.to_datetime(mege_df.Date)
mege_test_df['Date']=pd.to_datetime(mege_test_df.Date)
```

```python
mege_df['Day']=mege_df.Date.dt.day
mege_df['Month']=mege_df.Date.dt.month
mege_df['Year']=mege_df.Date.dt.year

mege_test_df['Day']=mege_test_df.Date.dt.day
mege_test_df['Month']=mege_test_df.Date.dt.month
mege_test_df['Year']=mege_test_df.Date.dt.year
```

```python
input_cols=['Store','DayOfWeek','Promo', 'StateHoliday','Year','Month','Day','Assortment','StoreType']
target_col=['Sales']
numeric_cols=['Store','Year','Month','Day']
categorical_cols=['DayOfWeek','Promo', 'StateHoliday','Assortment','StoreType']
```

```python
#mege_test_df_bk=mege_test_df.copy()
#mege_df_bk=mege_df.copy()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean').fit(mege_df[numeric_cols])
mege_df[numeric_cols]=imputer.transform(mege_df[numeric_cols])
mege_test_df[numeric_cols]=imputer.transform(mege_test_df[numeric_cols])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(mege_df[numeric_cols])
mege_df[numeric_cols] = scaler.transform(mege_df[numeric_cols])
mege_test_df[numeric_cols] = scaler.transform(mege_test_df[numeric_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(max_categories=5, sparse_output=False)


encoder.fit(mege_df[categorical_cols])

encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
mege_df[encoded_cols] = encoder.transform(mege_df[categorical_cols])
mege_test_df[encoded_cols] = encoder.transform(mege_test_df[categorical_cols])
```

```python
print(encoded_cols)
```

```python
from xgboost import XGBRegressor
?XGBRegressor
```

```python
model=XGBRegressor(random_state=42, n_jobs=-1, n_estimators=20, max_depth=3)
```

```python
train_targets=mege_df[target_col].copy()
X_train=mege_df[numeric_cols+ encoded_cols]
Y_test=mege_test_df[numeric_cols+ encoded_cols]

```

```python
model.fit(X_train, train_targets)
```

```python
from xgboost import plot_tree
plot_tree(model, rankdir='LR', num_trees=0)
```

```python
preds=model.predict(Y_test)
```

```python
preds
```

```python


submission_df=pd.read_csv('./rossmann-store-sales/sample_submission.csv')
submission_df['Sales']=preds*mege_test_df['Open'].astype('float')
submission_df.fillna(0,inplace=True)
submission_df.to_csv('submission_XGBoost.csv', index=None)
!head submission.csv
from IPython.display import FileLink
FileLink('submission.csv')
```
