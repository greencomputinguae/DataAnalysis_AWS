1. Download and prepare the dataset for training

```python
!pip install opendatasets --upgrade --quiet
import os
import pandas as pd
import opendatasets as od
#dataset_url = 'https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques'
#od.download(dataset_url)
data_dir = './house-prices-advanced-regression-techniques'
os.listdir(data_dir)
```

```python
train_csv = data_dir + '/train.csv'
train_df = pd.read_csv(train_csv)

test_csv = data_dir + '/test.csv'
test_df = pd.read_csv(test_csv)
```

```python
train_df
```

```python
train_df.describe()
```

```python
from sklearn.model_selection import train_test_split
input_cols, target_col = train_df.columns[1:-1], train_df.columns[-1]
```

```python
inputs_df, targets = train_df[input_cols].copy(), train_df[target_col].copy()
```

```python
import numpy as np
# Detect the numerical and categorical columns in dataframe
numeric_cols = train_df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = train_df.select_dtypes('object').columns.tolist()
```

```python
len(categorical_cols)+len(numeric_cols) #Dividing to numerical and categorical columns successfully 1460 rows × 81 columns
train_df.head(1)
```

```python
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.heatmap(train_df[numeric_cols].corr(), cmap='Reds', annot=True)
plt.title('Correlation Matrix')
```

```python
train_df[numeric_cols].corr() # it seems, there is not any strong correlation between SalePrice and other numerical features
```

```python
train_df[categorical_cols].nunique()
```

```python
train_df=inputs_df.copy()
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(handle_unknown='ignore')
encoder.fit(train_df[categorical_cols])
encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
print(encoded_cols)
train_inputs = train_df.copy()
test_inputs  = test_df.copy()
train_inputs[encoded_cols] = encoder.transform(train_df[categorical_cols]).toarray()
test_inputs[encoded_cols] = encoder.transform(test_df[categorical_cols]).toarray()
```

```python
train_inputs.head(1)
```

```python
train_inputs.isna().sum() 
```

```python
#finding columns with Nan values
#for (columnName, columnData) in train_inputs.iteritems():
#    if train_inputs[columnName].isna().sum():
#        print('Column Name : ', columnName)
#
numeric_cols_test=[item for item in numeric_cols if item!='SalePrice'] 
numeric_cols_test=[item for item in numeric_cols_test if item!='Id'] 
```

```python

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean')
imputer.fit(train_inputs[numeric_cols_test])
train_inputs[numeric_cols_test] = imputer.transform(train_inputs[numeric_cols_test])

```

```python
#numeric_cols = train_df.select_dtypes(include=np.number).columns.tolist()

#numeric_cols_test=numeric_cols
#numeric_cols_test.remove('SalePrice')

```

```python
#numeric_cols_test=[item for item in numeric_cols if item!='SalePrice'] #remove target from the numeric_columns
imputer.fit(test_inputs[numeric_cols_test])
test_inputs[numeric_cols_test] = imputer.transform(test_inputs[numeric_cols_test])
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler().fit(train_inputs[numeric_cols_test])
train_inputs[numeric_cols_test]= scaler.transform(train_inputs[numeric_cols_test])
```

```python
scaler = MinMaxScaler().fit(test_inputs[numeric_cols_test])
test_inputs[numeric_cols_test]= scaler.transform(test_inputs[numeric_cols_test])
```

```python
#train_inputs.drop(columns='Id')
#test_inputs.drop(columns=['Id'])
train_inputs.head(1)
```

```python
from sklearn import tree
clf=tree.DecisionTreeClassifier()
clf=clf.fit(train_inputs[numeric_cols_test+encoded_cols], targets)
Predications=clf.predict(test_inputs[numeric_cols_test+encoded_cols])
```
