```python
train_size=int(0.8*len(mege_df))
```

```python
sorted_df=mege_df.sort_values('Date')
train_df, val_df=sorted_df[:train_size], sorted_df[train_size:]
```

```python
len(train_df), len(val_df)
```

```python
input_cols=['Store','DayOfWeek','Promo', 'StateHoliday','Year','Month','Day','Assortment','StoreType']
target_col=['Sales']
```

```python
mege_df[input_cols].nunique()
```

```python
train_inputs=train_df[input_cols].copy()
train_targets=train_df[target_col].copy()

val_inputs=val_df[input_cols].copy()
val_targets=val_df[target_col].copy()
```

```python
numeric_cols=['Store','Year','Month','Day']
categorical_cols=['DayOfWeek','Promo', 'StateHoliday','Assortment','StoreType']

test_inputs=mege_test_df[input_cols].copy()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = 'mean').fit(train_inputs[numeric_cols])
train_inputs[numeric_cols]=imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols]=imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols]=imputer.transform(test_inputs[numeric_cols])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(train_inputs[numeric_cols])
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False, handle_unknown='ignore')
encoder.fit(train_inputs[categorical_cols])

encoded_cols = list(encoder.get_feature_names_out(categorical_cols))
```

```python
print(encoded_cols)
#adding sparse=False could cause columns must be same length as key ERROR
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categorical_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categorical_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categorical_cols])
```

```python
X_train=train_inputs[numeric_cols+ encoded_cols]
X_val=val_inputs[numeric_cols+ encoded_cols]
X_test=test_inputs[numeric_cols+ encoded_cols]
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
linreg=LinearRegression()
linreg.fit(X_train, train_targets)
```

```python
train_preds=linreg.predict(X_train)
mean_squared_error(train_preds, train_targets, squared=False)
```

```python
def try_model(model):
    #Fit the model
    model.fit(X_train, train_targets)
    
    #Generate predication
    train_preds=model.predict(X_train)
    val_preds=model.predict(X_val)
    
    #Computer loss (RMSE)
    train_rmse=mean_squared_error(train_preds, train_targets, squared=False)
    val_rmse  = mean_squared_error(val_preds, val_targets, squared=False)
    return train_rmse, val_rmse
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, SGDRegressor
```

```python
try_model(LinearRegression())
```

```python
try_model(Ridge())
```

```python
try_model(Lasso())
```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
for val in [32, 42, 52]:
    tree=DecisionTreeRegressor(random_state=val)
    print(try_model(tree))
```

```python
def predict_input(model, single_input):
    if single_input['Open']==0:
        return 0
    input_df=pd.DataFrame([single_input])
    input_df['Date']=pd.to_datetime(input_df.Date)
    input_df['Day']=pd.to_datetime(input_df.Date.dt.day)
    input_df['Month']=pd.to_datetime(input_df.Date.dt.month)
    input_df['Year']=pd.to_datetime(input_df.Date.dt.year)
    input_df[numeric_cols]=imputer.transform(input_df[numeric_cols])
    input_df[numeric_cols]=scaler.transform(input_df[numeric_cols])
    input_df[encoded_cols]=encoder.transform(input_df[categorical_cols])
    X_input=input_df[numeric_cols+encoded_cols]
    pred=model.predict(X_input)
    return pred
```
